```python
from transformers import T5Tokenizer
from datasets import DatasetDict

# Load tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-base")

# Max lengths (you can increase later)
MAX_INPUT_LENGTH = 512
MAX_TARGET_LENGTH = 128

# Step 1: Add 'summarize:' prefix
def preprocess_function(batch):
    input_texts = ["summarize: " + doc for doc in batch["article"]]
    target_texts = batch["highlights"]

    model_inputs = tokenizer(
        input_texts,
        max_length=MAX_INPUT_LENGTH,
        padding="max_length",
        truncation=True,
    )

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            target_texts,
            max_length=MAX_TARGET_LENGTH,
            padding="max_length",
            truncation=True,
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs


# Apply to all splits (train, validation)
tokenized_dataset = dataset.map(
    preprocess_function,
    batched=True,
    remove_columns=dataset["train"].column_names,
    desc="Tokenizing dataset",
)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer
import numpy as np
import math
import time
from tqdm import tqdm

# Constants
MAX_TARGET_LENGTH = 128
BATCH_SIZE = 16
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
VOCAB_SIZE = tokenizer.vocab_size
PAD_ID = tokenizer.pad_token_id
```

```python
# Subsample
train_encoded = tokenized_dataset["train"].select(range(20000))
eval_encoded = tokenized_dataset["validation"].select(range(25))

# PyTorch-compatible dataset
class SummarizationDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings["input_ids"])

    def __getitem__(self, idx):
        return {
            "input_ids": torch.tensor(self.encodings["input_ids"][idx]),
            "attention_mask": torch.tensor(self.encodings["attention_mask"][idx]),
            "labels": torch.tensor(self.encodings["labels"][idx]),
        }

train_dataset = SummarizationDataset(train_encoded)
eval_dataset = SummarizationDataset(eval_encoded)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
eval_loader = DataLoader(eval_dataset, batch_size=1)
```

```python
class Seq2SeqGRU(nn.Module):
    def __init__(self, vocab_size, embed_dim=512, hidden_dim=512, num_layers=1):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.encoder = nn.GRU(embed_dim, hidden_dim, num_layers, batch_first=True)
        self.decoder = nn.GRU(embed_dim, hidden_dim, num_layers, batch_first=True)
        self.fc_out = nn.Linear(hidden_dim, vocab_size)

    def forward(self, input_ids, labels):
        embedded_inputs = self.embedding(input_ids)
        _, hidden = self.encoder(embedded_inputs)

        embedded_labels = self.embedding(labels)
        decoder_outputs, _ = self.decoder(embedded_labels, hidden)
        logits = self.fc_out(decoder_outputs)
        return logits
```

```python
def train_epoch(model, dataloader, optimizer):
    model.train()
    total_loss = 0
    for batch in tqdm(dataloader, desc="Training"):
        input_ids = batch["input_ids"].to(DEVICE)
        labels = batch["labels"].to(DEVICE)

        optimizer.zero_grad()
        logits = model(input_ids, labels[:, :-1])  # teacher forcing
        loss = F.cross_entropy(
            logits.view(-1, VOCAB_SIZE),
            labels[:, 1:].reshape(-1),
            ignore_index=PAD_ID
        )
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)
```

```python
def greedy_decode(model, input_ids, max_length=MAX_TARGET_LENGTH):
    model.eval()
    batch_size = input_ids.size(0)

    decoder_input = torch.full((batch_size, 1), PAD_ID, dtype=torch.long).to(DEVICE)
    hidden = None

    with torch.no_grad():
        embedded_inputs = model.embedding(input_ids)
        _, hidden = model.encoder(embedded_inputs)

        outputs = []

        for _ in range(max_length):
            embedded = model.embedding(decoder_input[:, -1:])  # last predicted token
            output, hidden = model.decoder(embedded, hidden)
            logits = model.fc_out(output)
            next_token = logits.argmax(-1)
            decoder_input = torch.cat([decoder_input, next_token], dim=1)
            outputs.append(next_token)

        outputs = torch.cat(outputs, dim=1)  # remove initial pad
    return outputs
```

```python
def evaluate_model(model, dataloader):
    model.eval()
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    bleu_scores, rouge1s, rouge2s, rougeLs = [], [], [], []
    total_time = 0

    for batch in tqdm(dataloader, desc="Evaluating"):
        input_ids = batch["input_ids"].to(DEVICE)
        labels = batch["labels"].to(DEVICE)

        start_time = time.time()
        pred_ids = greedy_decode(model, input_ids, max_length=MAX_TARGET_LENGTH)
        total_time += time.time() - start_time

        for pred, tgt in zip(pred_ids, labels):
            pred_text = tokenizer.decode(pred, skip_special_tokens=True)
            tgt_text = tokenizer.decode(tgt, skip_special_tokens=True)

            bleu_scores.append(sentence_bleu([tgt_text.split()], pred_text.split()))

            scores = scorer.score(tgt_text, pred_text)
            rouge1s.append(scores["rouge1"].fmeasure)
            rouge2s.append(scores["rouge2"].fmeasure)
            rougeLs.append(scores["rougeL"].fmeasure)

    return {
        "bleu": np.mean(bleu_scores),
        "rouge1": np.mean(rouge1s),
        "rouge2": np.mean(rouge2s),
        "rougeL": np.mean(rougeLs),
        "latency_per_sample_sec": total_time / len(dataloader)
    }
```

```python
model = Seq2SeqGRU(vocab_size=VOCAB_SIZE).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Train one epoch
train_loss = train_epoch(model, train_loader, optimizer)
print(f"Train Loss: {train_loss:.4f}")

# Evaluate
metrics = evaluate_model(model, eval_loader)
print("ğŸ“Š Evaluation Metrics:")
for k, v in metrics.items():
    print(f"{k}: {v:.4f}")
```
